```python
interim_data_dir = Path(r'D:\Projects\0017. object detection ai ml engineer\object_detection\data\interim\coco')
with open(interim_data_dir / 'anno_coco.json', 'r') as f:
    coco_data = json.load(f)

class_count = {}
print(coco_data["categories"])
print(coco_data["annotations"][0]["category_id"])
for ann in coco_data["annotations"]:
    category_id = ann["category_id"]
    cat_name = coco_data["categories"][category_id-1]["name"]
    class_count[cat_name] = class_count.get(cat_name, 0) + 1
print(class_count)
```

```python
plt.bar(class_count.keys(), class_count.values())
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Distribution of Classes in COCO Dataset")
plt.xticks(rotation=45)
plt.show()
```
